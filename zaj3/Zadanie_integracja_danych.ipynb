```python
df_co2 = pd.read_csv("co2-emissions-per-capita.csv")
df_co2.head()
```

```python
df_primary_energy = pd.read_csv("primary-energy-cons.csv")
df_primary_energy.head()
```

```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

```python
co2_pl = df_co2[df_co2["Entity"] == "Poland"]
energy_pl = df_primary_energy[df_primary_energy["Entity"] == "Poland"]
```

```python
poland = pd.merge(
    co2_pl,
    energy_pl,
    on="Year",
    how="inner"
)
```

```python
poland_final = poland[[
    "Year",
    "Annual CO₂ emissions (per capita)",
    "Primary energy consumption (TWh)"
]].rename(columns={
    "Annual CO₂ emissions (per capita)": "co2_per_capita",
    "Primary energy consumption (TWh)": "primary_energy_TWh"
})
```

```python
poland_final.to_csv("poland_energy_co2.csv", index=False)
```

```python
countries = {
    "UE": "European Union (27)",
    "USA": "United States",
    "Chiny": "China",
    "Indie": "India"
}

with pd.ExcelWriter("energy_co2.xlsx", engine="openpyxl") as writer:
    for sheet, country in countries.items():
        co2_c = df_co2[df_co2["Entity"] == country]
        energy_c = df_primary_energy[df_primary_energy["Entity"] == country]

        merged = pd.merge(co2_c, energy_c, on="Year", how="inner")

   
        merged_clean = merged[[
            "Year",
            "Annual CO₂ emissions (per capita)",
            "Primary energy consumption (TWh)"
        ]].rename(columns={
            "Annual CO₂ emissions (per capita)": "co2_per_capita",
            "Primary energy consumption (TWh)": "primary_energy_TWh"
        })

        merged_clean.to_excel(writer, sheet_name=sheet, index=False)
```
