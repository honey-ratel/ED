ZADANIE 1: Analiza danych sprzedażowych
---------------------------------------
Dane: Miesięczna sprzedaż produktu w PLN

sprzedaz = [12500, 13200, 15800, 14200, 16500, 18200, 17800, 19200,
            20500, 21300, 23500, 25200]

a) Oblicz podstawowe statystyki (średnia, mediana, min, max)

b) Znajdź miesiące ze sprzedażą powyżej średniej

c) Oblicz procentowy wzrost sprzedaży (grudzień vs styczeń)

d) Stwórz wykres liniowy pokazujący trend

e) Dodaj linię średniej na wykresie


````python
def oblicz_statystyki(dane):
    """
    Oblicza podstawowe statystyki opisowe dla listy liczb.

    Parameters:
    -----------
    dane : list
        Lista wartości numerycznych

    Returns:
    --------
    dict : Słownik ze statystykami
    """
    return {
        "srednia": sum(dane) / len(dane),
        "mediana": sorted(dane)[len(dane)//2],
        "min": min(dane),
        "max": max(dane),
        "n": len(dane)
    }
````

````python
sprzedaz = [12500, 13200, 15800, 14200, 16500, 18200, 17800, 19200, 20500, 21300, 23500, 25200]
wynik = oblicz_statystyki(sprzedaz)
````

````python
wynik["srednia"]
````

````python
wynik["mediana"]
````

````python
wynik["min"]
````

````python
wynik["max"]
````

````python
#Znajdź miesiące ze sprzedażą powyżej średniej
print("Miesiace w ktorych sprzedarz była powyżej średniej:")
for i, x in enumerate(sprzedaz):
    if x > wynik["srednia"]:
        print(i+1)
````

````python
#Oblicz procentowy wzrost sprzedaży (grudzień vs styczeń)
wzrost_procentowy = (sprzedaz[-1] - sprzedaz[0]) / sprzedaz[0] * 100
wzrost_procentowy
````

````python
#d) Stwórz wykres liniowy pokazujący trend
x = np.linspace(1,12, len(sprzedaz))
y = sprzedaz

plt.figure(figsize=(10, 4))
plt.plot(x, y, 'b-', linewidth=2, label='Sprzedarz')
plt.xlabel('Miesiac')
plt.ylabel('Sprzedaz')
plt.title('Trend sprzedarzy')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
````

````python
plt.plot(x, y, 'b-', linewidth=2, label='Sprzedarz')
plt.plot(x, [wynik["srednia"]]*len(x), 'r--', linewidth=2, label='srednia')
````

ADANIE 2: Słowniki - Analiza portfolio
---------------------------------------
Stwórz słownik opisujący portfolio inwestycyjne:
```
portfolio = {
    'AAPL': {'liczba': 10, 'cena_zakupu': 150, 'cena_aktualna': 175},
    'GOOGL': {'liczba': 5, 'cena_zakupu': 2800, 'cena_aktualna': 2950},
    'MSFT': {'liczba': 15, 'cena_zakupu': 300, 'cena_aktualna': 320}
}
```

a) Oblicz całkowitą wartość początkową portfolio

b) Oblicz aktualną wartość portfolio

c) Oblicz zysk/stratę dla każdej akcji

d) Które akcje przynoszą największy zwrot (w %)?


````python
portfolio = {
    'AAPL': {'liczba': 10, 'cena_zakupu': 150, 'cena_aktualna': 175},
    'GOOGL': {'liczba': 5, 'cena_zakupu': 2800, 'cena_aktualna': 2950},
    'MSFT': {'liczba': 15, 'cena_zakupu': 300, 'cena_aktualna': 320}
}
````

````python
wartosc_pocz = 0;
for dane in portfolio.values():
    wartosc_pocz += dane['liczba'] * dane['cena_zakupu']

print(wartosc_pocz)
````

````python
wartosc_akt = 0;
for dane in portfolio.values():
    wartosc_akt += dane['liczba'] * dane['cena_aktualna']

print(wartosc_akt)
````

````python
zysk_strata = {}
for akcja, dane in portfolio.items():
    koszt = dane['liczba'] * dane['cena_zakupu']
    zysk =  dane['liczba'] * dane['cena_aktualna']
    zysk_strata[akcja] = zysk - koszt
print(zysk_strata)
````

````python
zysk_strata = {}
for akcja, dane in portfolio.items():
    koszt = dane['liczba'] * dane['cena_zakupu']
    zysk =  dane['liczba'] * dane['cena_aktualna']
    zysk_strata[akcja] = (zysk - koszt) / koszt * 100
print(zysk_strata)
````

miesiace = ['Sty', 'Lut', 'Mar', 'Kwi', 'Maj', 'Cze',

c) Oblicz średnią różnicę temperatur między latami

````python
temp_2022 = [2, 4, 8, 13, 18, 21, 24, 23, 19, 14, 8, 3]
temp_2023 = [3, 5, 9, 14, 19, 22, 25, 24, 20, 15, 9, 4]
miesiace = ['Sty', 'Lut', 'Mar', 'Kwi', 'Maj', 'Cze', 'Lip', 'Sie', 'Wrz', 'Paź', 'Lis', 'Gru']
````

````python

plt.plot(miesiace, temp_2022, marker='o', label='2022')
plt.plot(miesiace, temp_2023, marker='o', label='2023')
max_2022 = max(temp_2022)
min_2022 = min(temp_2022)
max_2023 = max(temp_2023)
min_2023 = min(temp_2023)
plt.plot(miesiace[temp_2022.index(max_2022)], max_2022, marker='o', markersize=10)
plt.plot(miesiace[temp_2022.index(min_2022)], min_2022, marker='o', markersize=10)
````

````python
suma_2022 = 0
suma_2023 = 0
for temp in temp_2022:
    suma_2022 += temp
for temp in temp_2023:
    suma_2023 += temp
print("srednia roznica miedzy latami 2022 i 2023 ")
print((suma_2022/len(temp_2022)) - (suma_2023/len(temp_2023)))
````

ZADANIE 4: Symulacja Monte Carlo
---------------------------------
Zasymuluj 1000 rzutów kostką (6-ścienną):

a) Wygeneruj dane używając np.random.randint(1, 7, 1000)

b) Stwórz histogram pokazujący częstość każdej liczby

c) Oblicz teoretyczne i empiryczne prawdopodobieństwo każdej liczby

d) Czy wyniki są zgodne z teorią? (każda liczba ~16.67%)

````python
rzuty = np.random.randint(1, 7, 1000)
````

````python
plt.figure()
plt.hist(rzuty, bins=[1, 2, 3, 4, 5, 6, 7], rwidth=0.8)
plt.xticks([1, 2, 3, 4, 5, 6])
plt.show()
````

````python
teoretyczne_p = 1 / 6
print("Teoretyczne prawdopodobieństwo:")
print(teoretyczne_p)

print("Empiryczne prawdopodobieństwa:")
for i in range(1, 7):
    empiryczne = np.sum(rzuty == i) / len(rzuty)
    print(empiryczne)
````
