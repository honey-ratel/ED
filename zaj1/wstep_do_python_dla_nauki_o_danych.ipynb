```python
## 3.6 Przykład: Agregacja danych
# Liczba wystąpień każdej temperatury
from collections import Counter
licznik_temp = Counter(dane)
print(f"\nRozkład temperatur: {dict(licznik_temp)}")
```

```python
## 3.7 Mapowanie kategorii (przydatne w data science!)
kategorie_temp = {
    "zimno": (0, 15),
    "chlodno": (15, 20),
    "umiarkowanie": (20, 25),
    "cieplo": (25, 30),
    "goraco": (30, 100)
}

def kategoryzuj(temp):
    for kategoria, (min_t, max_t) in kategorie_temp.items():
        if min_t <= temp < max_t:
            return kategoria
    return "nieznana"

temp_test = 23
print(f"\n{temp_test}°C to: {kategoryzuj(temp_test)}")
```

# CZĘŚĆ 4: FUNKCJE

```python
def oblicz_statystyki(dane):
    """
    Oblicza podstawowe statystyki opisowe dla listy liczb.

    Parameters:
    -----------
    dane : list
        Lista wartości numerycznych

    Returns:
    --------
    dict : Słownik ze statystykami
    """
    return {
        "srednia": sum(dane) / len(dane),
        "mediana": sorted(dane)[len(dane)//2],
        "min": min(dane),
        "max": max(dane),
        "n": len(dane)
    }
```

```python
# Test funkcji
wyniki = oblicz_statystyki(temperatury)
print("Statystyki temperatur:")
for k, v in wyniki.items():
    print(f"  {k}: {v:.2f}")
```

```python
## Funkcja z wartością domyślną
def standaryzuj(dane, srednia=None, std=None):
    """Standaryzacja danych (z-score)"""
    if srednia is None:
        srednia = sum(dane) / len(dane)
    if std is None:
        wariancja = sum((x - srednia)**2 for x in dane) / len(dane)
        std = wariancja ** 0.5

    return [(x - srednia) / std for x in dane]
```

```python
dane_norm = standaryzuj([1, 2, 3, 4, 5])
print(f"\nDane znormalizowane: {[round(x, 2) for x in dane_norm]}")
```

# CZĘŚĆ 7: Rozkłady statystyczne

## 7.1 Rozkład normalny (Gaussa)

```python
def gauss_pdf(x, mu, sigma):
    """Funkcja gęstości prawdopodobieństwa rozkładu normalnego"""
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)

x = np.linspace(-5, 9, 1000)

plt.figure(figsize=(12, 6))
plt.plot(x, gauss_pdf(x, 0, 1), 'b-', linewidth=2, label='μ=0, σ=1')
plt.plot(x, gauss_pdf(x, 0, 2), 'r-', linewidth=2, label='μ=0, σ=2')
plt.plot(x, gauss_pdf(x, 2, 1), 'g-', linewidth=2, label='μ=2, σ=1')
plt.xlabel('x', fontsize=12)
plt.ylabel('Gęstość prawdopodobieństwa', fontsize=12)
plt.title('Rozkład normalny dla różnych parametrów', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 7.2 Porównanie teoretycznego i empirycznego rozkładu

```python
## 7.2 Porównanie teoretycznego i empirycznego rozkładu
np.random.seed(42)
dane_empiryczne = np.random.normal(0, 1, 10000)

plt.figure(figsize=(12, 6))
plt.hist(dane_empiryczne, bins=50, density=True, alpha=0.7,
         color='skyblue', edgecolor='black', label='Dane empiryczne')
x_teoria = np.linspace(-4, 4, 200)
plt.plot(x_teoria, gauss_pdf(x_teoria, 0, 1), 'r-',
         linewidth=3, label='Rozkład teoretyczny')
plt.xlabel('Wartość', fontsize=12)
plt.ylabel('Gęstość prawdopodobieństwa', fontsize=12)
plt.title('Porównanie rozkładu empirycznego i teoretycznego', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
