````python
data_spam = {
    'Słowo_1': ['darmowy', 'darmowy', 'spotkanie', 'raport', 'oferta', 'darmowy',
                'spotkanie', 'oferta', 'raport', 'darmowy', 'spotkanie', 'oferta'],
    'Słowo_2': ['wygrana', 'wygrana', 'jutro', 'kwartalny', 'specjalna', 'rabat',
                'dziś', 'limitowana', 'miesięczny', 'rabat', 'pilne', 'wyjątkowa'],
    'Wykrzyknik': ['TAK', 'TAK', 'NIE', 'NIE', 'TAK', 'TAK',
                   'NIE', 'TAK', 'NIE', 'TAK', 'NIE', 'TAK'],
    'Spam': ['TAK', 'TAK', 'NIE', 'NIE', 'NIE', 'TAK',
             'NIE', 'NIE', 'NIE', 'TAK', 'NIE', 'NIE']
}
````

````python
import pandas as pd

df = pd.DataFrame(data_spam)
````

````python
p_spam = (df["Spam"] == "TAK").mean()
p_nospam = (df["Spam"] == "NIE").mean()
````

````python
def conditional_prob(feature, value, spam_class):
    subset = df[df["Spam"] == spam_class]
    return (subset[feature] == value).mean()


new_email = {
    "Słowo_1": "darmowy",
    "Słowo_2": "wygrana",
    "Wykrzyknik": "TAK"
}
````

````python
p1 = p_spam
for feature, value in new_email.items():
    p1 *= conditional_prob(feature, value, "TAK")


p0 = p_nospam
for feature, value in new_email.items():
    p0 *= conditional_prob(feature, value, "NIE")
````

````python
print("Spam:", p1)
print("Nie Spam:", p0)
````
